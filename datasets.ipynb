```python
# get counts for each ingredient
# parse each row of ingredients into a list from a string representation of list
all_ingredients = set()

for ingredients_string in raw_recipes["ingredients"]:
    ingredients_list = ingredients_string.strip('][').split(', ')
    for ingredient in ingredients_list:
        all_ingredients.add(ingredient.strip())

len(all_ingredients)
```

```python
# show counts for each individual ingredient

from collections import Counter
from tqdm import tqdm
# use threadpoolexecutor to speed up the process
from concurrent.futures import ThreadPoolExecutor

counts = Counter()

def count_ingredient(review):
    for ingredient in all_ingredients:
        if ingredient in str(review):
            counts[ingredient] += 1

with ThreadPoolExecutor(8) as executor:
    executor.map(count_ingredient, tqdm(raw_interactions['review'][:50000], total=len(raw_interactions)))

counts
```
